## Why People Analytics

Data is everywhere and make meaning out of it can be extremely complicated. People Analytics provides tools, methodologies and techniques to extract meaning out of employee data and ***make this data useful*** to:

- Interpret of large volume of data;
- Identify trends and patterns in employee data;
- Help predicting organizations' and employees needs;
- Prioritize HR activities based on impact utility and return on investment;
- Reduce subjectivity and make decision-making transparent.

:::footer

# Business problem discovery
In this phase there are a few primarly goals:

- Identify Problem
- Define objectives and success metrics
- Determine data sources and team 


## Identify problem
- What is the problem?
- Is it a problem? What's the business value?
- Who are the stakeholders involved?
    - direct
    - indirect
- Scope
- Time frame

## Define objectives
We normally try to answer these types of questions with People Data Science:

- How much? (regression)
- Which category? (classification)
- Which group? (clustering)
- Which option should be taken? (recommendation system)

## Data Acquisition
Data selection refers to the process of collecting, retrieving, gathering, and sourcing data. 

![](AdobeStock_119801832-edited-scaled.jpeg){.absolute .fragment bottom="0" right="200" width="700"}


## Subjective data
- **Surveys**
    - *Pros*: affordable, familiar to employees and (if well designed!!) very effective
    - *Cons*: bias-sensitive and (possibly) could induce fatigue  
- **Performance reviews and rating forms**
    - *Pros*: efficient and (potentially) informative
    - *Cons*K: bias-sensitive and (often) unreliable

## Objective
- **Surveillance and monitoring**
    - *Pros*: objective behaviour measures, and (potentially) rich.
    - *Cons*: intrusive, costly to store

- **Organisational information** 
    - **Pros** Often available and cheap
    - **Cons** data quality variable and should be well documented

- **Scraped data**
    - **Pros**: Rich and new tools facilitate extraction
    - **Cons**: Ethica/privacy issues and require lots of pre-processing
